### 1.Condicional: If Else

**Note:** Verifica a equivalência de dois valores


```python
var_n1 = 2
var_n2 = 0

if var_n2 == var_n1 and var_n1 != 0:
    print("Os números são equivalentes")
elif var_n2 == 0:
    print("var_n2 é igual a zero")
else:
    print("Os números são divergentes")
```

### 2. Laço de Repetição

**Note:** Loop For In

```python
tupla_numeros = (1,2,3,4,5,6)

for item in tupla_numeros:
    if item%2 == 0:
        print("Item:", item, " é par")
    else:
        print("Item:", item, " é impar")
```

**Note:** Loop For In Usando Rage

```python
for aux in range(0,5):
    print(aux)
```

**Note:** While

```python
var = 0
while var < 5:
    print(var)
    var = var + 1
else:
    print("Loop finalizado !")
```

### 2.Método

**Note:** A função DIR exibe todos os médotos para um objeto

```python
list =[]
dir(list)
```

**Note:** Criando uma função

```python
def insert_number(first_number, second_number):
    print("Primeiro Número:",first_number)
    print("Segundo Número:",second_number)
    print("Soma dos Números:",(first_number + second_number))
```

**Note:** Efetuando uma chamada de função

```python
insert_number(1,2)
```

### 3.Expressões Lambdas

**Note**: Não utiliza a palavra reservada "Def" e o comando fica na mesma linha

```python
lambda_par = lambda num:  num%2==0
lambda_par(2)
```

```python
lambda_soma = lambda n1, n2 : n1 + n2
lambda_soma(1,1)
```

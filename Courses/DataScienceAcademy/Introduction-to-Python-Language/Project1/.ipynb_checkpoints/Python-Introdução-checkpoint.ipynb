### 5.Strings

**Note:** As strings podem ser atribuidas usando aspas SIMPLES e as DUPLAS

```python
var_teste = 'Olá mundo ! [1]'
print(var_teste)
var_teste = "Olá mundo ! [2]"
print(var_teste)
```

**Note:** Realizando a impressão da primeira letra de String

```python
var_teste[0]
```

**Note:** "Fatiando" uma String

```python
print("Retornando String a partir da posição (1)",var_teste[1:])
print("Retornando String até a posição (3)",var_teste[:3])
print("Retornando último Caracter",var_teste[-1])
print("Retornando String a partir do penultimo caracter",var_teste[:-1])
print("Fatiando String de 2 em 2 Caracteres",var_teste[::2]) #O padrão é 1
```

**Note:** Concatenando Strings

```python
nome = "Jefferson"
sobrenome = "Alves"
nome_completo = nome + " " + sobrenome

nome_completo
```

**Note:** Multiplicando Strings

```python
var_teste = "w";
var_teste * 3
```

### 6.Listas

**Note:** Criando uma Lista

```python
list_linguagem_de_programacao = ["python","java","C#"]
list_linguagem_de_programacao
```

**Note:** Criando uma Lista com diferentes tipos

```python
meus_dados = ["Jefferson Alves", 1.70, 11]
meus_dados
```

**Note:** Atualizando item da lista

```python
meus_dados[0] = "Jefferson Alves Santos"
meus_dados
```

**Note:** Deletando item da lista

```python
del meus_dados[1] #Usando Índice
print(meus_dados)

meus_dados.remove("Jefferson Alves Santos") #Usando String
print(meus_dados)
```

**Note:** Listas Alinhadas

```python
lista_alinhada = [[1,2,3],[4,5,6]]
print("Exibindo Todos os Dados da Lista Alinahda: ", lista_alinhada)
print("Exibindo Lista de Índice 1 da Lista Alinahda: ", lista_alinhada[1])
print("Exibindo Lista de Indice 1 da lista alinhada e Dado do Indice 1 da Lista: ",lista_alinhada[1][1])
```

**Note:** Efetuando consultas em listas

```python
5 in lista_alinhada[1]
```

**Note:** Funções Básicas para Listas

```python
lista = [15,20,1]
print("Valor Máximo da Lista: ", max(lista))
print("Valor Mínimo da Lista: ", min(lista))
print("Contando a Lista: ", len(lista))
print("Contando a Lista (com base no parametro): ", lista.count("15"))
print("Consultando Índice da Lista: ", lista.index(20))

#Adicionando Item ao final da lista
lista.append(2) 
lista
```

### 7.Dicionários


**Note:** Declarando um Dicionário

```python
dicionario_cliente = {"nome":"josé","idade":30,"estado_civil":"solteiro"}
dicionario_cliente
```

**Note:** Funções Básicas para Dicionairios

```python
dicionario_cliente.pop("nome","josé")
print(dicionario_cliente)
print(dicionario_cliente.keys())
print(dicionario_cliente.values())
print(dicionario_cliente.items())
```
